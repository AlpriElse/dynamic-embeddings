```python
STM_FILE_PATH = '../datasets/TEDLIUM_release-3/data/stm'
OUTPUT_FILE_PATH = '../datasets/TEDLIUM_release-3/data/raw'

stm_files = os.listdir(STM_FILE_PATH)

# stm_files = stm_files[0:2] # temp
# stm_files
```

```python
def remove_prefixed_metadata(transcript_line):
        start_index = transcript_line.find('<NA>') + len('<NA>')
        return transcript_line[start_index:]

def convert_file(filename):
    with open(f'{STM_FILE_PATH}/{filename}') as content_file:
        content = content_file.read()
        lines = content.split('\n')
        
        raw_lines = list(map(remove_prefixed_metadata, lines))
        output_text = '\n'.join(raw_lines)

        with open(f'{OUTPUT_FILE_PATH}/{filename[:-4]}.txt', 'w+') as output_file:
            output_file.write(output_text)
            output_file.close()
```

```python
for i, filename in enumerate(stm_files):
    print(i, filename)
    convert_file(filename)
```
